# Separating Features and Label(target)

```python
train_df=df[df['COST'] !=0.0]
train_df.shape
```

```python
test_df=df[df['COST']==0.0]
test_df.drop('COST',axis=1,inplace=True)
test_df.shape
```

```python
x=train_df.drop('COST',axis=1)
y=train_df["COST"].values
```

```python
x_test=test_df
```

# Modelling or Model Building

```python
#importing necessary libraries
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor,ExtraTreesRegressor
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor as KNN
```

```python
# finding the best Random State
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_cv,y_train,y_cv = train_test_split(x,y,test_size=.20, random_state=i)
    lr=LinearRegression()
    lr.fit(x_train, y_train)
    pred = lr.predict(x_cv)
    acc=r2_score(y_cv, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print('Maximum r2 score is',maxAccu,'on Random_state',maxRS)
```

```python
LR = LinearRegression()
LR.fit(x_train,y_train)
pred_LR=LR.predict(x_cv)
pred_train=LR.predict(x_train)
print('R2_score:',r2_score(y_cv,pred_LR))
print('R2_score on training Data:',r2_score(y_train,pred_train)*100)
print('Mean Absolute Error:-',mean_absolute_error(y_cv, pred_LR))
print('Mean Squared Error:-',mean_squared_error(y_cv, pred_LR))
print('Root Mean Squared Error:-',np.sqrt(mean_squared_error(y_cv, pred_LR)))
score = cross_val_score(LR,x,y, cv=5, scoring='r2')
print(score)
print(score.mean())
print('Difference between R2 score and cross validation score is =', (r2_score(y_cv,pred_LR) - score.mean())*100)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=pred_LR,y=y_cv,color='r')
plt.plot(pred_LR,pred_LR,color='b')
plt.xlabel('Actual',fontsize=14)
plt.ylabel('Predicted', fontsize=14)
plt.title('Linear Regression', fontsize=18)
plt.show()
```

```python
RFR=RandomForestRegressor()
RFR.fit(x_train,y_train)
pred_RFR=RFR.predict(x_cv)
pred_train=RFR.predict(x_train)
print('R2_score:',r2_score(y_cv,pred_RFR))
print('R2_score on training Data:',r2_score(y_train,pred_train)*100)
print('Mean Absolute Error:-',mean_absolute_error(y_cv, pred_RFR))
print('Mean Squared Error:-',mean_squared_error(y_cv, pred_RFR))
print('Root Mean Squared Error:-',np.sqrt(mean_squared_error(y_cv, pred_RFR)))
score1 = cross_val_score(RFR,x,y)
print(score1)
print(score1.mean())
print('Difference between R2 score and cross validation score is – ', (r2_score(y_cv,pred_RFR) - score1.mean())*100)
```

```python
knn=KNN()
knn.fit(x_train,y_train)
pred_knn=knn.predict(x_cv)
pred_train=knn.predict(x_train)
print('R2_score:',r2_score(y_cv,pred_knn))
print('R2_score on training Data:',r2_score(y_train,pred_train)*100)
print('Mean Absolute Error:-',mean_absolute_error(y_cv, pred_knn))
print('Mean Squared Error:-',mean_squared_error(y_cv, pred_knn))
print('Root Mean Squared Error:-',np.sqrt(mean_squared_error(y_cv, pred_knn)))
score2 = cross_val_score(knn,x,y)
print(score2)
print(score2.mean())
print('Difference between R2 score and cross validation score is =', (r2_score(y_cv,pred_knn) - score2.mean())*100)
```

```python
GBR=GradientBoostingRegressor()
GBR.fit(x_train,y_train)
pred_GBR=GBR.predict(x_cv)
pred_train=GBR.predict(x_train)
print('R2_score:',r2_score(y_cv,pred_GBR))
print('R2_score on training Data:',r2_score(y_train,pred_train)*100)
print('Mean Absolute Error:-',mean_absolute_error(y_cv, pred_GBR))
print('Mean Squared Error:-',mean_squared_error(y_cv, pred_GBR))
print('Root Mean Squared Error:-',np.sqrt(mean_squared_error(y_cv, pred_GBR)))
score3 = cross_val_score(GBR,x,y)
print(score3)
print(score3.mean())
print('Difference between R2 score and cross validation score is –', (r2_score(y_cv,pred_GBR) - score3.mean())*100)
```
